```python
import math
```

```python
pisos = [10, 100, 100, 1000, 1000]
tazas = [2, 2, 3, 3, 5]
```

```python
def comprobando_los_pisos(k, tazas):
    if tazas == 2:
        return k + (k * (k-1)) // 2

    elif tazas == 3:
        return k + (k * (k-1)) // 2 + (k * (k-1) * (k-2)) // 6

    elif tazas == 5:
        return k + (k * (k-1)) // 2 + (k * (k-1) * (k-2)) // 6 + (k * (k - 1) * (k - 2) * (k - 3)) // 24 + (k * (k - 1) * (k - 2) * (k - 3) * (k - 4)) // 120
    
    else:
        return None
```

```python
def construir_escalera(pisos_max, k):
    pasos = []
    actual = 0
    paso = k

    while actual < pisos_max and paso > 0:
        actual += paso
        pasos.append(min(actual, pisos_max))
        paso -= 1

    return pasos
```

```python
print("+------------+------------+------------------+-------------------------------------------+")
print("| Поверхи    | Стакани    | Спроби (k)       | Кроки                                     |")
print("+------------+------------+------------------+-------------------------------------------+")

for i in range(len(pisos)):

    p = pisos[i]
    t = tazas[i]

    k = 1
    while comprobando_los_pisos(k, t) < p:
        k += 1

    if t == 2:
        pasos = construir_escalera(p, k)
        estrategia = ", ".join(map(str, pasos))
    else:
        estrategia = "no lineal"
    
    print(f"| {p:<10} | {t:<10} | {k:<16} | {estrategia:<41} |")

print("+------------+------------+------------------+-------------------------------------------+")
```
